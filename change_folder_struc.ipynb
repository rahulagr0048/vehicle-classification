```python
from glob import glob
import pandas as pd
from shutil import copyfile
```

```python
files = glob('{}/*/*_image.jpg'.format('trainval'))
files.sort()
```

```python
print(len(files))
files
```

```python
df = pd.read_csv('trainval/labels.csv')
```

```python
len(files)
```

```python
i = 0;
for file in files:
    guid = file.split('/')[-2]
    idx = file.split('/')[-1].replace('_image.jpg', '')
    if (guid + "/" + idx == df['guid/image'][i]):
        src = "trainval/" + guid + "/" + idx + "_image.jpg"
        if (df['label'][i] == 0):
            dst = "train/0/" + guid + idx + "_image.jpg"
        elif (df['label'][i] == 1):
            dst = "train/1/" + guid + idx + "_image.jpg"
        else:
            dst = "train/2/" + guid + idx + "_image.jpg"
    copyfile(src, dst)
    i = i + 1;
```
